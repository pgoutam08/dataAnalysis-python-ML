```python
import pandas as pd
hr = pd.read_csv(r"C:\Users\pattargv\Downloads\Test_Data\HR_salary.csv")
hr.head(5)
```

```python
hr["Department"].unique()
```

```python
rating_dict = {'Very Good': 6,'Good':5, 'Average':4,'Poor':3, 'Very Poor':2, 'Not Rated':1}
dept_dict = {'Accounting':1,'Accounts': 1, 'Business Development': 2, 'Engineering': 3, 'Human Resources': 4, 'Legal': 5,
             'Marketing':6, 'Product Management': 7, 'Research and Development':8, 'Sales':9, 'Services': 10, 'Support':11, 
             'Training':12}
gender_dict = {'Male':1,'Female': 2, 'Not Revealed': 3}


hr['Rating'] = hr['Rating'].map(rating_dict)
hr['Department'] = hr['Department'].map(dept_dict)
hr['Gender'] = hr['Gender'].map(gender_dict)

hr.head()
```

```python
import sklearn
from sklearn import svm

#hr = hr.drop("Name", axis = 1)
hr.head()
```

```python
X = hr.drop("Salary", axis =1 ).values
#X = preprocessing.scale(X)
y = hr["Salary"].values



test_size = 200


X_train = X[:-test_size]
y_train = y[:-test_size]


X_test = X[-test_size:]
y_test = y[-test_size:]
```

```python
clf = svm.SVR(kernel = "linear")
clf.fit(X_train, y_train)
```

```python
print(clf.score(X_test, y_test))

for X,y in zip(X_test, y_test):
    print(f"Model: {clf.predict([X])[0]}, Actual: {y}")
```

```python
clf1 = svm.SVR(kernel = "rbf")
clf1.fit(X_train, y_train)
```

```python
print(clf1.score(X_test, y_test))

for X,y in zip(X_test, y_test):
    print(f"Model: {clf1.predict([X])[0]}, Actual: {y}")
```

```python
df = pd.read_csv(r"C:\Users\pattargv\Downloads\Test_Data\diamonds.csv", index_col=0)
df.head()
```

```python
df['cut'].unique()
```

```python
cut_class_dict = {"Fair": 1, "Good": 2, "Very Good": 3, "Premium": 4, "Ideal": 5}
clarity_dict = {"I3": 1, "I2": 2, "I1": 3, "SI2": 4, "SI1": 5, "VS2": 6, "VS1": 7, "VVS2": 8, "VVS1": 9, "IF": 10, "FL": 11}
color_dict = {"J": 1,"I": 2,"H": 3,"G": 4,"F": 5,"E": 6,"D": 7}

df['cut'] = df['cut'].map(cut_class_dict)
df['clarity'] = df['clarity'].map(clarity_dict)
df['color'] = df['color'].map(color_dict)
```

```python
df.head()
```

```python
df.tail()
```

```python
import sklearn
from sklearn import svm, preprocessing

df = sklearn.utils.shuffle(df)

X = df.drop("price", axis = 1).values
X = preprocessing.scale(X)

y = df["price"].values

```

```python
test_size = 200

X_trail = X[:-test_size] 
y_trail = y[:-test_size]


X_test = X[-test_size:]
y_test = y[-test_size:]


clf = svm.SVR(kernel="linear")
clf.fit(X_trail, y_trail)
```

```python
print(clf.score(X_test, y_test))

for X,y in zip(X_test, y_test):
    print(f"Model: {clf.predict([X])[0]}, Actual: {y}")
```

```python
clf = svm.SVR(kernel="rbf")
clf.fit(X_trail, y_trail)


print(clf.score(X_test, y_test))

for X,y in zip(X_test, y_test):
    print(f"Model: {clf.predict([X])[0]}, Actual: {y}")
```
